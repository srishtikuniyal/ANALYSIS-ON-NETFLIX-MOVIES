HANDLING THE NULL VALUES

```python
dataset['seasons'].fillna(round(dataset['seasons'].mean(),0), inplace=True)
dataset['imdb_score'].fillna(dataset['imdb_score'].mean(), inplace=True)
dataset['imdb_votes'].fillna(dataset['imdb_votes'].mean(), inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset['age_certification'].fillna(dataset['age_certification'].mode()[0], inplace=True)
dataset['imdb_id'].fillna(dataset['imdb_id'].mode()[0], inplace=True)
dataset['title'].fillna(dataset['title'].mode()[0], inplace=True)
```

DATA VISUALIZATION

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset['imdb_score'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of IMDb Ratings for Netflix Movies')
plt.xlabel('IMDb Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the number of movies by release year
plt.figure(figsize=(14, 6))
sns.countplot(x='release_year', data=dataset, palette='viridis')
plt.title('Number of Netflix Movies by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot IMDb ratings over the years
plt.figure(figsize=(14, 6))
sns.lineplot(x='release_year', y='imdb_score', data=dataset, ci=None, marker='o', color='orange')
plt.title('IMDb Ratings Over the Years for Netflix Movies')
plt.xlabel('Release Year')
plt.ylabel('IMDb Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
type_counts = dataset['type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('Distribution of Movies and Shows on Netflix')
plt.show()
```

```python
age_certification_counts = dataset['age_certification'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(age_certification_counts, labels=age_certification_counts.index,autopct='%1.1f%%', colors=['skyblue', 'lightcoral','red','yellow'], startangle=180)
plt.title('Distribution of Movies and Shows on Netflix')
plt.show()
```

```python
from scipy.stats import norm

plt.figure(figsize=(15,7))
sns.distplot(dataset['runtime'], color='red')
plt.title('Distplot with Normal distribution for Movies',fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x = dataset["genres"].value_counts().head(10).index,
            y = dataset["genres"].value_counts().head(10).values,palette="pink")
plt.xticks(rotation=60)
plt.title("Top10 Genre in Movies",fontweight="bold")
plt.show()
```
